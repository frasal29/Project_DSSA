```python
# Definisco la classe attività, per una rappresentazione ad oggetti, con attributi: nome, durata, lista attività precedenti e successive
class Attivita:
  def __init__(self, nome, durata):
    self.nome_attivita = nome
    self.durata_attivita = durata
    self.lista_successori = []
    self.lista_predecessori = []

  def aggiungi_successore(self, attivita_successiva):
    self.lista_successori.append(attivita_successiva)
    attivita_successiva.lista_predecessori.append(self)


  def aggiungi_predecessore(self, attivita_precedente): #utilizzato in particolare per il nodo pozzo
    self.lista_predecessori.append(attivita_precedente)
    attivita_precedente.lista_successori.append(self)
```

```python
# Funzione che prende in in input il numero di attività e restituisce una lista con la durata di ogni singola attività
def lista_attivita(numero_attivita):
  lista_durata_attivita = []

  print("Ora dovrai fornirmi la durata di ogni attività\n")

  for i in range(numero_attivita):
    print("Che durata ha l'attivita", i+1)
    while True:
      try:
        durata_att = float(input())
        lista_durata_attivita.append(durata_att)
        break
      except:
        print("\nInserisci un intero o un numero decimale")
        print("Che durata ha l'attivita", i+1)

  return lista_durata_attivita



```

```python
# Funzione che prende in in input il numero di attività e restituisce una lista con la durata casuale (float da 0 a 10) di ogni singola attività
import random
def lista_attivita_casuale(numero_attivita):
  lista_durata_attivita = []

  # Attribuzione casuale delle durate delle attività da 0 a 10
  for i in range(numero_attivita):
    durata_att_random = round(random.uniform(0, 10),2) # genera un numero float casuale da 0 a 10. Round arrotonda il numero con 2 cifre decimali
    lista_durata_attivita.append(durata_att_random)

  print("Bene! Ho attribuito casualmente la durata ad ogni attività, in un range da 0 a 10\n")
  return lista_durata_attivita
```

```python
# Funzione che dato in input il nome di una attivita la cerca nella lista di oggetti delle attività
# e restituisce l'elemento (oggetto), solo se questo è un intero
def cerca_nome_attivita(lista_oggetti, nome_att):
  if isinstance(nome_att, int) is False: # se nome_att non è un intero restituisci None
    return None
  for i, elemento in enumerate(lista_oggetti):
    if elemento.nome_attivita == nome_att:
      return elemento

```

```python
# Funzione che mi dice quante sono il numero di attività successive rispetto all'attivita attuale,
# nella lista di sole attività interne, cioè senza i nodi s e t
def numero_attivita_possibili(lista, attivita):
  # Si da' per scontato che l'attività è presente nella lista.
  # Devo farmi restituire la posizione nella lista di questa attivita
  posizione_nella_lista = 0
  for i, elemento in enumerate(lista):
    if elemento.nome_attivita == attivita:
      posizione_nella_lista = i

  # Se la posizione è l'ultima della lista vuol dire che non c'è nessuna attività possibile successiva quindi ritorno 0
  if posizione_nella_lista == len(lista)-1:
    return 0

  # Ora ho la posizione nella lista e devo scorrerla a partire da quella posizione +1, fino alla fine
  conteggio = 0
  for elem in lista[posizione_nella_lista+1:]:
    conteggio += 1
  return conteggio

```

```python
# Funzione che data la lista dei successori di una attivita e il nome di una attivita mi restituisce True
# se è presente il nome nella lista, False altrimenti
def attivita_is_in_lista_successori(lista_successori, attivita):
  for i, elem in enumerate(lista_successori):
    if elem.nome_attivita == attivita:
      return True
  return False
```

```python
#funzione che stampa il grafo con durata di ogni nodo e successori
def stampa_grafo(lista_oggetti_attivita):
  for i, elemento in enumerate(lista_oggetti_attivita):
    print("Attività: ", elemento.nome_attivita)
    print("Durata: ", elemento.durata_attivita)
    print("Successori: ")

    for j, elem in enumerate(elemento.lista_successori):
      print("", elemento.nome_attivita, " --> ", elem.nome_attivita, "\n")

    print("\n\n")
```

```python
#funzione che assegna ad ogni nodo ogni successore
def assegna_successore(lista_oggetti_attivita):
  print("Ora bisogna assegnare ad ogni attività, le attività successive, cioè quelle che vengono eseguite dopo quella attuale.")
  print("Il nodo sorgente ha durata 0 ed è collegato a tutti i nodi del grafo, mentre il nodo terminale ha anch'esso durata 0\ne tutti i nodi del grafo sono collegati ad esso")

  #collego il nodo s a tutti gli altri nodi. Esso è il primo della lista_oggetti_attivita
  nodo_sorgente = lista_oggetti_attivita[0]
  for elemento in lista_oggetti_attivita[1:-1]: #ciclo su tutte le attivita tranne nodo sorgente e nodo pozzo
    nodo_sorgente.aggiungi_successore(elemento)

  #collego tutti gli altri nodi (escluso s) al nodo t. Esso è l'ultimo della lista_oggetti_attivita
  nodo_pozzo = lista_oggetti_attivita[-1] # l'ultimo elemento ha indice -1
  for elemento in lista_oggetti_attivita[1:-1]: #ciclo su tutte le attivita tranne nodo sorgente e nodo pozzo
    nodo_pozzo.aggiungi_predecessore(elemento) # in questo modo anche tutti i nodi (tranne s e t) hanno come successore t


  #settaggio successori nodi interni (tutti tranne s e t)
  for i, elemento in enumerate(lista_oggetti_attivita[1:-1]): # ciclo su tutti i nodi interni
    nome_att = elemento.nome_attivita
    attivita_successive_massimo = numero_attivita_possibili(lista_oggetti_attivita[1:-1], nome_att) # salvo il numero massimo di attività possibili come successori

    if attivita_successive_massimo == 0: # se le attività successori massime possibili sono pari a 0, vuol dire che era l'ultimo nodo interno, quindi esci dal ciclo
      print("L'attività ", nome_att, "non può avere attività successive\n")
      break

    print("\nDimmi la quantità di attività successive all'attività ", nome_att, " . Esse non possono essere superiori a: ", attivita_successive_massimo, "")
    while True:
      try:
        numero_att = int(input())
        break
      except:
        print("Inserisci un intero!")

    while numero_att > attivita_successive_massimo or numero_att < 0:
      print("Scegli un numero di attività successive all'attività", nome_att, "plausibile")
      print("Dimmi la quantità di attività successive all'attività ", nome_att, " . Esse non possono essere superiori a: ", attivita_successive_massimo, "")
      while True:
        try:
          numero_att = int(input())
          break
        except:
          print("Inserisci un intero!")
          print("Scegli un numero di attività successive all'attività", nome_att, "plausibile")
          print("Dimmi la quantità di attività successive all'attività ", nome_att, " . Esse non possono essere superiori a: ", attivita_successive_massimo, "")

    if numero_att == 0: # se viene scelto un numero pari a 0 di attività successive, passa al prossimo nodo
      print("Bene! L'attività ", nome_att, " non ha nessuna attività successiva")
      continue

    for x in range(numero_att): # cicla sul numero di attività successive
      while True:
        try:
          print("Dimmi il nome dell'attività successiva a ", nome_att)
          att_succ = int(input())
          break
        except:
          print("Devi inserire un intero!")

      oggetto_att_succ = cerca_nome_attivita(lista_oggetti_attivita[1:-1], att_succ) # vedi se è presente questa attività successiva nel grafo

      while oggetto_att_succ is None or att_succ <= nome_att or attivita_is_in_lista_successori(elemento.lista_successori, att_succ) is True:
        print("L'attività da te scelta o non è presente nel progetto o ha un numero minore rispetto a quella attuale\no è l'attività corrente stessa, o è stata già inserita come attività successiva")
        while True:
          try:
            print("Dimmi il nome dell'attività successiva a ", nome_att)
            att_succ = int(input())
            break
          except:
            print("Inserisci un intero!")
        oggetto_att_succ = cerca_nome_attivita(lista_oggetti_attivita[1:-1], att_succ)

      print("Bene! Hai inserito correttamente l'attività successiva. La direzione è questa:", nome_att, " -> ", att_succ, "")
      elemento.aggiungi_successore(oggetto_att_succ) # aggiungi l'oggetto corrispondente all'attività successiva nella lista dei successori del nodo




```

```python
import random
#funzione che assegna ad ogni nodo ogni successore
def assegna_successore_casuale(lista_oggetti_attivita):
  print("Ora assegno casualmente ad ogni attività, le attività successive, cioè quelle che vengono eseguite dopo quella attuale.")
  print("Il nodo sorgente ha durata 0 ed è collegato a tutti i nodi del grafo, mentre il nodo terminale ha anch'esso durata 0\ne tutti i nodi del grafo sono collegati ad esso")

  #collego il nodo s a tutti gli altri nodi. Esso è il primo della lista_oggetti_attivita
  nodo_sorgente = lista_oggetti_attivita[0]
  for elemento in lista_oggetti_attivita[1:-1]: #ciclo su tutte le attivita tranne nodo sorgente e nodo pozzo
    nodo_sorgente.aggiungi_successore(elemento)

  #collego tutti gli altri nodi (escluso s) al nodo t. Esso è l'ultimo della lista_oggetti_attivita
  nodo_pozzo = lista_oggetti_attivita[-1] # l'ultimo elemento ha indice -1
  for elemento in lista_oggetti_attivita[1:-1]: #ciclo su tutte le attivita tranne nodo sorgente e nodo pozzo
    nodo_pozzo.aggiungi_predecessore(elemento) # in questo modo anche tutti i nodi (tranne s e t) hanno come successore t


  #settaggio casuale successori nodi interni (tutti tranne s e t e penultimo nodo)
  for i, elemento in enumerate(lista_oggetti_attivita[1:-1]):
    nome_att = elemento.nome_attivita
    attivita_successive_massimo = numero_attivita_possibili(lista_oggetti_attivita[1:-1], nome_att) # massime attività possibili come successori all'elemento corrente

    if attivita_successive_massimo == 0: # se le attività successori massime possibili sono pari a 0, vuol dire che era l'ultimo nodo interno, quindi esci dal ciclo
      print("L'attività ", nome_att, "non può avere attività successive\n")
      break

    print("\nLa quantità di attività successive all'attività ", nome_att, " non può essere superiore a: ", attivita_successive_massimo, "\n")

    numero_att = random.randint(0, attivita_successive_massimo) # scelgo un numero intero di attività successive tra 0 e il massimo
    print("Ho assegnato all'attività", nome_att, " un numero di attività successive pari a ", numero_att, "\n")

    if numero_att == 0: # se viene assegnato 0 passa al prossimo nodo
      print("L'attività ", nome_att, " non ha nessuna attività successiva perché gli è stata assegnata una quantità di attività successive pari a 0")
      continue

    for x in range(numero_att): # cicla sul numero di attività successive scelto
      att_succ = random.randint(int(nome_att)+1, int(lista_oggetti_attivita[-2].nome_attivita)) # sceglie una attività successiva tra quella con nome +1 e la penultima
      while attivita_is_in_lista_successori(elemento.lista_successori, att_succ) is True: # se è stata già scelta questa attività successiva
        print("Cerco attività successiva...\n")
        att_succ = random.randint(int(nome_att)+1, int(lista_oggetti_attivita[-2].nome_attivita)) # ripete l'operazione di scelta attività successiva

      print("Bene! Ho inserito correttamente l'attività successiva. La direzione è questa:", nome_att, " -> ", att_succ, "")
      oggetto_att_succ = cerca_nome_attivita(lista_oggetti_attivita[1:-1], att_succ) # risale all'oggetto del nome attivita successiva
      elemento.aggiungi_successore(oggetto_att_succ) # inserisce l'oggetto dell'attività successiva alla lista successori dell'elemento corrente




```

```python
# Funzione che prende in input la lista delle attivita (cioè quella della durata di ogni attivita) e restituisce una lista di oggetti
# in cui ogni oggetto è un nodo della classe attivita
def costruttore_lista_oggetti_attivita(lista_durata_att):
  lista_oggetti_attivita = []

  # Costruisco il nodo sorgente con nome 's' e durata 0
  nodo_sorgente = Attivita('s', 0)
  lista_oggetti_attivita.append(nodo_sorgente)

  # Sulla base delle durate delle attivita nella lista in argomento costruisco gli oggetti e li inserisco in una lista, cioè gli oggetti sono i nodi interni
  for indice, elemento in enumerate(lista_durata_att):
    nodo = Attivita(indice+1, elemento) # costruisco l'oggetto partendo dal num 1, in cui elemento è la sua durata
    lista_oggetti_attivita.append(nodo) # inserisco l'oggetto nella lista

  # Costruisco il nodo pozzo con nome 't' e durata 0
  nodo_pozzo = Attivita('t', 0)
  lista_oggetti_attivita.append(nodo_pozzo)

  print("\nIl progetto è composto dalle seguenti attività:\n")
  for i, elemento in enumerate(lista_oggetti_attivita):
    print("Attivita:", lista_oggetti_attivita[i].nome_attivita)
    print("Durata:", lista_oggetti_attivita[i].durata_attivita)
    print("\n")

  return lista_oggetti_attivita
```

```python
# Funzione che calcola la durata dei cammini tramite una ricerca in profondità DFS (Deep First Search)
def dfs_calcola_durata(attuale, nodo_destinazione, durata_corrente, cammino_corrente, visitati):
    # Se il nodo attuale è il nodo destinazione, restituisci la durata corrente e il cammino.
    if attuale == nodo_destinazione:
        return [(durata_corrente, cammino_corrente)]

    risultati = []  # Lista di tuple per memorizzare le durate e i cammini.
    visitati.add(attuale)  # Segna il nodo attuale come visitato.

    # Visita tutti i successori dell'attività corrente che non sono ancora stati visitati.
    for successore in attuale.lista_successori:
        if successore not in visitati: # se questo successore non è stato visitato
            # Chiamata ricorsiva
            risultati += dfs_calcola_durata(successore, nodo_destinazione, durata_corrente + successore.durata_attivita, cammino_corrente + [successore.nome_attivita], visitati.copy())

    return risultati

# Funzione che salva in risultati tutte le durate dei cammini e restituisce la durata massima
def durata_cammini_da_s_a_t(lista_oggetti_attivita):
    nodo_sorgente = lista_oggetti_attivita[0]
    nodo_pozzo = lista_oggetti_attivita[-1]
    risultati = dfs_calcola_durata(nodo_sorgente, nodo_pozzo, 0, [nodo_sorgente.nome_attivita], set())

    # Trova il cammino con la durata massima
    durata_massima = max(risultati, key=lambda x: x[0]) # restituisce il massimo della durate, considerando solo i primi elementi delle tuple, che sono proprio le durate di ogni cammino

    return durata_massima
```

```python
import networkx as nx # per la rappresentazione di grafi
import matplotlib.pyplot as plt # per mostrare il grafo

# Funzione che prende in input la lista oggetti delle attività e il cammino_max, e disegna il grafo con i cammino massimo in rosso
def from_object_to_digraph(lista_ogg_attivita, cammino_max):
  digrafo = nx.DiGraph() # crea un nuovo grafo orientato vuoto
  for nodo in lista_ogg_attivita: # ciclo sugli oggetti attività
    id = nodo.nome_attivita
    durata = nodo.durata_attivita
    successori = nodo.lista_successori
    for succ in successori: # per tutti i successori del nodo corrente
      # Aggiungo un arco ponderato (con peso pari alla durata dell'attività)
      # al grafo orientato tra il nodo attuale e ciascuno dei suoi successori
      digrafo.add_weighted_edges_from([(id, succ.nome_attivita, durata)])

  pos = nx.kamada_kawai_layout(digrafo)  # algoritmo per una migliore posizione nello spazio dei nodi
  pos['s'] = [-1, 0.5] # metto nodo s tutto a sx e metà altezza
  pos['t'] = [1, 0.5] # metto nodo t tutto a dx e metà altezza

  # Crea liste per i colori degli archi e dei nodi
  edge_colors = []
  node_colors = []

  # Assegna colori agli archi
  for edge in digrafo.edges(): # itera su tutti gli archi
    # Se i due nodi di questo arco fanno parte del cammino e se sono consecutivi, colora l'arco di rosso
    if edge[0] in cammino_max and edge[1] in cammino_max and cammino_max.index(edge[0]) + 1 == cammino_max.index(edge[1]):
      edge_colors.append('red')  # colore per gli archi nel cammino
    # Altrimenti, colora l'arco di grigio
    else:
      edge_colors.append('gray')  # colore per tutti gli altri archi

  # Assegna colori ai nodi
  for node in digrafo.nodes(): # itera su tutti i nodi
    # Se il nodo fa parte del cammino max coloralo di rosso
    if node in cammino_max:
      node_colors.append('red')  # colore rosso per i nodi nel cammino
    # Altrimenti, coloralo di azzurro chiaro
    else:
      node_colors.append('lightblue')  # colore azzurro per tutti gli altri nodi

  # Disegna il grafo con i colori assegnati di nodi e archi. Gli archi e nodi del cammino sono rossi
  nx.draw(digrafo, pos, with_labels=True, node_color=node_colors, edge_color=edge_colors, font_weight="bold") # bold è il font in grassetto

  labels = nx.get_edge_attributes(digrafo, 'weight') # estrate i pesi degli archi

  # Disegna i pesi degli archi con un font leggibile e piccolo
  nx.draw_networkx_edge_labels(digrafo, pos, edge_labels=labels, font_size=8, bbox=dict(facecolor='none', edgecolor='none', boxstyle='round,pad=0.1'))

  plt.title("GRAFO DELLE ATTIVITA, CON DURATA DEL PROGETTO (CAMMINO ROSSO)") # titolo grafico
  plt.show() # mostra grafico

```

```python
#MAIN
print("Ciao sono un programma che calcola il tempo necessario per la durata di un progetto\n")
print("Dimmi il numero di attività di questo progetto\n")
while True:
  try:
    num_attivita = int(input())
    if num_attivita > 0:
      break
    else:
      print("Per favore, inserisci un numero maggiore di 0.")
  except:
    print("Per favore, inserisci un numero intero valido.")


print("Bene! Ora vuoi attribuire a ciascuna attività una durata (premere tasto 1),\noppure genero casualmente una durata per ciascuna attività (premere tasto 2) ?")
while True:
  try:
    scelta = int(input())
    if scelta == 1:
      print("Hai fatto la scelta")
      lista_att = lista_attivita(num_attivita)
      lista_ogg_attivita = costruttore_lista_oggetti_attivita(lista_att)
      assegna_successore(lista_ogg_attivita)
      stampa_grafo(lista_ogg_attivita)
      durata_max, cammino_max = durata_cammini_da_s_a_t(lista_ogg_attivita) # scompone la tupla in durata_max e cammino_max
      print("Durata massima del progetto:", durata_max)
      print("Cammino con durata massima:", " -> ".join(str(attivita) for attivita in cammino_max))
      from_object_to_digraph(lista_ogg_attivita, cammino_max)
      break

    if scelta == 2:
      print("Hai fatto la scelta")
      lista_att = lista_attivita_casuale(num_attivita)
      lista_ogg_attivita = costruttore_lista_oggetti_attivita(lista_att)
      assegna_successore_casuale(lista_ogg_attivita)
      stampa_grafo(lista_ogg_attivita)
      durata_max, cammino_max = durata_cammini_da_s_a_t(lista_ogg_attivita) # scompone la tupla in durata_max e cammino_max
      print("Durata massima del progetto:", durata_max)
      print("Cammino con durata massima:", " -> ".join(str(attivita) for attivita in cammino_max))
      from_object_to_digraph(lista_ogg_attivita, cammino_max)
      break

    else:
      print("Devi scegliere l'intero 1 oppure 2")

  except:
    print("Devi inserire un intero: il numero 1 oppure il numero 2\n")

print("Fine!")
```
